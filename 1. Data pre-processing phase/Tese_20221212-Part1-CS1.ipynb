```python
import file
import ipywidgets as widgets

path = file.gui_fname()
path = path.decode('latin-1')
print("Filename: ", path)
header = 0

def f(x):
    return x

x=widgets.IntSlider(min=0, max=5, step=1, value=0, description="Header:")
x
```

```python
import pandas as pd
import numpy as np
import dateutil.parser
import datetime
import os

name, extension = os.path.splitext(path)
header = x.value

if "xls" in extension:
    data = pd.read_excel(path, header=header, sheet_name=None)
# data = pd.read_excel(r'Qnt prod__Dados_João.xlsx', header=0, sheet_name=None)
elif "csv" in extension:
    df = pd.read_csv(path, header=header)

if "xls" in extension:    
    df = pd.DataFrame()
    for key, value in data.items():
        df = df.append(pd.DataFrame(data[key]), ignore_index=True)
        
    df.rename(columns={'Unnamed: 0':'Componentes'}, inplace=True)
    
for k in df.keys():
    if not isinstance(k, str):
        df = df.rename(columns={k: str(k)})

display(df.head())
```

```python
def get_month(s):
    month_dict = {1: "jan", 2:"fev", 3:"mar", 4: "abr", 5:"mai", 6:"jun", 7: "jul", 8:"ago", 9:"set", 10: "out", 11:"nov", 12:"dec"}
    for month in month_dict:
        if month_dict[month] in s:
            return month
        
for date in df["Data"]:
    if not isinstance(date, datetime.date):
        if "/" in date:
            d = date.split("/")
            day, month, year = int(d[0]), int(d[1]), int(d[2])
#             print("day", day, "month", month, "year", year)
        else:
            print("date: ", date)
    else:
        day, month, year = date.day, date.month, date.year
    
    newdate = datetime.date(year=year,day=day,month=month)
    weekday = newdate.weekday()

df['Data'].replace(date, newdate, inplace=True)
df['Data'] = df['Data'].apply(pd.to_datetime)
df['day'] = pd.DatetimeIndex(df['Data']).day
df['month'] = pd.DatetimeIndex(df['Data']).month
df['year'] = pd.DatetimeIndex(df['Data']).year
df['week day'] = pd.DatetimeIndex(df['Data']).weekday


df['months'] = df['Data'].iloc[0]

print(df.dtypes)

df['months'] = ((df['Data'] - df['months'])/np.timedelta64(1, 'M')).astype(int)
display(df.head())
```
